**Biblioteki:** Requests, BeautifulSoup4  

```python
import requests

# przykład 1 - sprawdzenie dostępności serwera
# funkcja ger wysyła zapytanie do serwera
# status 200 oznacza sukces, a np. 404 błąd
response = requests.get("https://www.merito.pl/")
print(f"Status połączenia: {response.status_code}")

# przykład 2 - pobieranie kodu html strony
# .text zawiera cały kod html strony w formie stringa
# dla przykładu wycinamy pierwsze 1000 znaków
response = requests.get("https://www.merito.pl/")
print(response.text[:1000])
```

```python
from bs4 import BeautifulSoup

# przykład 1 - wyodrębnienie konkretnych tagów
# zmienna html symuluje kod
# html.parser to narzędzie, które "rozumie" strukturę tagów
# wyciągamy sam tytuł strony
html = "<html><head><title>Przykład 1</title></head><body><p>Witaj!</p></body></html>"
soup = BeautifulSoup(html, 'html.parser')

print(f"Tytuł strony to: {soup.title.string}")

# przykład 2 - wyszukiwanie wielu elementów
# .find_all('a') znajduje wszystkie tagi <a> (linki) umieszczone w kodzie
# .get('href') pozwala wyciągnąć wartość atrybutu "href", czyli sam adres url
html = '<div><a href="https://www.merito.pl/szczecin/">Link 1</a><a href="https://www.merito.pl/poznan/">Link 2</a></div>'
soup = BeautifulSoup(html, 'html.parser')

print("Znalezione linki:")
for link in soup.find_all('a'):
    print(link.get('href'))
```
